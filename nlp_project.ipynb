```python
from textblob import TextBlob
```

```python
lang = ['arabic','chinese','english','spanish']
```

```python
data = {}
for i in range(1,11):
    data[i] = {}
    data[i]['name'] = open('pro2/'+str(i)+'/name.txt').read()
    data[i]['origin'] = open('prodata/'+str(i)+'/origin.txt').read()
    for l in lang:
        data[i][l] = open('pro2/'+str(i)+'/'+l+'.txt').read()
```

```python
for i in range(1,11):
    #print('Name of the movie: '+data[i]['name'])
    ptot=0
    stot=0
    for l in lang:
        #print(l,TextBlob(data[i][l]).sentiment)
        data[i][l+'_pol'] = TextBlob(data[i][l]).polarity
        ptot+=data[i][l+'_pol']
        data[i][l+'_sub'] = TextBlob(data[i][l]).subjectivity
        stot+=data[i][l+'_sub']
    ptot/=len(lang)
    stot/=len(lang)
    data[i]['mean_pol'] = ptot
    data[i]['mean_sub'] = stot
    #print('\n\n\n')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(columns = ['id','lang','pol','sub'])
```

```python
k=0
for i in range(1,11):
    for l in lang:
        df.loc[k] = [i,l,data[i][l+'_pol'],data[i][l+'_sub']]
        k+=1
    df.loc[k] = [i,'mean',data[i]['mean_pol'],data[i]['mean_sub']]
    k+=1
```

```python
arabic = df[df['lang']=='arabic']
arabic.reset_index(drop=True,inplace=True)
chinese = df[df['lang']=='chinese']
chinese.reset_index(drop=True,inplace=True)
english = df[df['lang']=='english']
english.reset_index(drop=True,inplace=True)
spanish = df[df['lang']=='spanish']
spanish.reset_index(drop=True,inplace=True)
mean = df[df['lang']=='mean']
mean.reset_index(drop=True,inplace=True)
```

```python
arabic.plot(x='id',y='pol',kind='bar',title='arabic')
chinese.plot(x='id',y='pol',kind='bar',title='chinese')
english.plot(x='id',y='pol',kind='bar',title='english')
spanish.plot(x='id',y='pol',kind='bar',title='spanish')
mean.plot(x='id',y='pol',kind='bar',title='mean')
plt.show()
```

```python
arabic['norm_pol'] = (arabic['pol']-mean['pol'])**2
chinese['norm_pol'] = (chinese['pol']-mean['pol'])**2
english['norm_pol'] = (english['pol']-mean['pol'])**2
spanish['norm_pol'] = (spanish['pol']-mean['pol'])**2
```

```python
arabic.plot(x='id',y='norm_pol',kind='bar',title='arabic')
chinese.plot(x='id',y='norm_pol',kind='bar',title='arabic')
english.plot(x='id',y='norm_pol',kind='bar',title='arabic')
spanish.plot(x='id',y='norm_pol',kind='bar',title='arabic')
```
