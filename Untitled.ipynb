```python
from textblob import TextBlob
```

```python
lang = ['chinese','english']
```

```python
data = {}
dsize=21
for i in range(dsize):
    data[i] = {}
    data[i]['name'] = open('prodata/'+str(i)+'/name.txt').read()
    data[i]['name']=data[i]['name'][:-1]
    data[i]['origin'] = open('prodata/'+str(i)+'/origin.txt').read()
    for l in lang:
        data[i][l] = open('prodata/'+str(i)+'/'+l+'.txt').read()
        data[i][l+'1'] = data[i][l][:int(len(data[i][l])/4)]
        data[i][l+'2'] = data[i][l][int(len(data[i][l])/4):int(len(data[i][l])/4)*2]
        data[i][l+'3'] = data[i][l][int(len(data[i][l])/4)*2:int(len(data[i][l])/4)*3]
        data[i][l+'4'] = data[i][l][int(len(data[i][l])/4)*3:]
```

```python
for i in range(dsize):
    print('Name of the movie: '+data[i]['name'])
    ptot1=0
    stot1=0
    ptot2=0
    stot2=0
    ptot3=0
    stot3=0
    ptot4=0
    stot4=0
    for l in lang:
        #print(l,TextBlob(data[i][l]).sentiment)
        data[i][l+'1_pol'] = TextBlob(data[i][l+'1']).polarity
        data[i][l+'2_pol'] = TextBlob(data[i][l+'2']).polarity
        data[i][l+'3_pol'] = TextBlob(data[i][l+'3']).polarity
        data[i][l+'4_pol'] = TextBlob(data[i][l+'4']).polarity
        ptot1+=data[i][l+'1_pol']
        ptot2+=data[i][l+'2_pol']
        ptot3+=data[i][l+'3_pol']
        ptot4+=data[i][l+'4_pol']
        
        data[i][l+'1_sub'] = TextBlob(data[i][l+'1']).subjectivity
        data[i][l+'2_sub'] = TextBlob(data[i][l+'2']).subjectivity
        data[i][l+'3_sub'] = TextBlob(data[i][l+'3']).subjectivity
        data[i][l+'4_sub'] = TextBlob(data[i][l+'4']).subjectivity
        stot1+=data[i][l+'1_sub']
        stot2+=data[i][l+'2_sub']
        stot3+=data[i][l+'3_sub']
        stot4+=data[i][l+'4_sub']
    ptot1/=len(lang)
    stot1/=len(lang)
    ptot2/=len(lang)
    stot2/=len(lang)
    ptot3/=len(lang)
    stot3/=len(lang)
    ptot4/=len(lang)
    stot4/=len(lang)
    data[i]['mean1_pol'] = ptot1
    data[i]['mean1_sub'] = stot1
    data[i]['mean2_pol'] = ptot2
    data[i]['mean2_sub'] = stot2
    data[i]['mean3_pol'] = ptot3
    data[i]['mean3_sub'] = stot3
    data[i]['mean4_pol'] = ptot4
    data[i]['mean4_sub'] = stot4
    #print('\n\n\n')
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(columns = ['id','lang','pol','sub','origin'])
```

```python
k=0
for i in range(dsize):
    for l in lang:
        df.loc[k] = [i,l,data[i][l+'1_pol'],data[i][l+'1_sub'],data[i]['origin']]
        k+=1
        df.loc[k] = [i,l,data[i][l+'2_pol'],data[i][l+'2_sub'],data[i]['origin']]
        k+=1
        df.loc[k] = [i,l,data[i][l+'3_pol'],data[i][l+'3_sub'],data[i]['origin']]
        k+=1
        df.loc[k] = [i,l,data[i][l+'4_pol'],data[i][l+'4_sub'],data[i]['origin']]
        k+=1
    df.loc[k] = [i,'mean',data[i]['mean1_pol'],data[i]['mean1_sub'],data[i]['origin']]
    k+=1
    df.loc[k] = [i,'mean',data[i]['mean2_pol'],data[i]['mean2_sub'],data[i]['origin']]
    k+=1
    df.loc[k] = [i,'mean',data[i]['mean3_pol'],data[i]['mean4_sub'],data[i]['origin']]
    k+=1
    df.loc[k] = [i,'mean',data[i]['mean3_pol'],data[i]['mean4_sub'],data[i]['origin']]
    k+=1
```
